# Waves

Physicists have found that nature tends to be very wavy, or seeing things as waves at times can drastically simplify equations and computations used to describe natural phenomena.

### Mathematical forms of wavelike behavior:

#### Trigonometric functions: 

Before we begin to further demonstrate these functions, we define a range of points that we want to be inputs for the function. In this particular situation, we are defining a range from $\theta = -4\pi \rightarrow -4\pi$ with a step size of .001 between the points between this range:

```python
theta = np.arange(-4.0*np.pi, 4.0*np.pi,.001)
```

$$ \mathrm{cos}(\theta) $$

```python
cosine_theta = np.cos(theta)
```

$$ \mathrm{sin}(\theta) $$

```python
sine_theta = np.sin(theta)
```

```python
# Plotting both sine and cosine functions:
plt.figure(figsize=[12.5,5])
ax = plt.axes()
ticks=np.arange(-3.5*np.pi,4.0*np.pi, np.pi/2.0)
ax.set_xticks(ticks)
ax.set_xticklabels([r'-$7/2 \pi$', r'-$ 3\pi$', r'-$5/2 \pi$', r'-$2\pi$', r'-$3/2 \pi$', r'-$\pi$', r'-$1/2 \pi$', r'0', r'$1/2 \pi$', r'$\pi$', r'$3/2 \pi$', r'$2\pi$', r'$5/2 \pi$', r'3$\pi$', r'$7/2 \pi$'])
plt.plot(theta, cosine_theta, label=r'$\mathrm{cos}(\theta)$')
plt.plot(theta, sine_theta, label=r'$\mathrm{sin}(\theta)$')
plt.legend()
plt.xlim(theta[0], theta[-1])
plt.ylabel('[arb]')
plt.xlabel('[rad]')
plt.grid(linestyle='dotted')
```

### Exponential functions (using Euler's equation):

$$e^{i\theta} = \mathrm{cos}(\theta) + i\mathrm{sin}(\theta) $$

$$ e^{-i\theta} = \mathrm{cos}(\theta) - i\mathrm{sin}(\theta) $$


Defining the first function from the two above on the same range we used before ($\theta = -4\pi \rightarrow -4\pi$)

```python
pos_exp = np.exp(1j*theta)
```

```python
# Plotting both sine and cosine functions:
plt.figure(figsize=[12.5,5])
ax = plt.axes()
ticks=np.arange(-3.5*np.pi,4.0*np.pi, np.pi/2.0)
ax.set_xticks(ticks)
ax.set_xticklabels([r'-$7/2 \pi$', r'-$ 3\pi$', r'-$5/2 \pi$', r'-$2\pi$', r'-$3/2 \pi$', r'-$\pi$', r'-$1/2 \pi$', r'0', r'$1/2 \pi$', r'$\pi$', r'$3/2 \pi$', r'$2\pi$', r'$5/2 \pi$', r'3$\pi$', r'$7/2 \pi$'])
plt.plot(theta, np.real(pos_exp), label=r'$Re(\mathrm{e}^{i\theta})$')
plt.plot(theta, np.imag(pos_exp), label=r'$Im(\mathrm{e}^{i\theta}$')
plt.legend()
plt.ylabel('[arb]')
plt.xlabel('[rad]')
plt.xlim(theta[0], theta[-1])
plt.grid(linestyle='dotted')
```

Look familiar? I hope it does because you actually plotted the same thing that you plotted a few cells back! Kind of cool especially since a single <u>*complex function*</u> has the ability to bundle information about both functions.
